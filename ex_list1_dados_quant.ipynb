```python
"""
1) Os dados hipotéticos apresentados referem-se às alturas (m) de 50 árvores
jovens de araucária:
a) Agrupar os dados em uma tabela de distribuição de frequências.
b) Fazer um histograma e polígono de frequências (baseados na
frequência absoluta).
c) Indique os valores que caracterizam a distribuição em relação aos
quartis (Q1, Q2 e Q3).

"""

import numpy as np
import matplotlib.pyplot as plt

# Dados das alturas das árvores
alturas = [
    2.81, 3.19, 3.49, 3.76, 6.02, 8.23, 2.23, 3.01, 4.43, 13.94,
    3.1, 1.52, 3.38, 2.85, 4.64, 7.33, 6.78, 13.12, 13.84, 9.4,
    6.2, 2.39, 9.19, 7.07, 9.2, 13.46, 3.9, 8.99, 7.97, 5.15,
    12.95, 25.52, 6.61, 16.56, 9.6, 6.71, 6.73, 3.86, 3.5, 4.8,
    8.4, 13.86, 6.53, 18.44, 22.14, 9.15, 8.75, 10.86, 14.2, 10.09
]

# Criar uma tabela de distribuição de frequências
bins = np.histogram_bin_edges(alturas, bins='auto')  # Calcular os bins automaticamente
hist, bin_edges = np.histogram(alturas, bins=bins)

# Criando o histograma e o polígono de frequências
fig, ax = plt.subplots(2, 1, figsize=(10, 12))

# Histograma
ax[0].hist(alturas, bins=bin_edges, color='skyblue', edgecolor='black')
ax[0].set_title('Histograma das Alturas das Araucárias')
ax[0].set_xlabel('Altura (m)')
ax[0].set_ylabel('Frequência Absoluta')

# Polígono de frequências
# Calcula o ponto médio de cada bin para o polígono de frequências
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])
ax[1].plot(bin_centers, hist, marker='o', linestyle='-', color='blue')
ax[1].set_title('Polígono de Frequências das Alturas das Araucárias')
ax[1].set_xlabel('Altura (m)')
ax[1].set_ylabel('Frequência Absoluta')

plt.tight_layout()
plt.show()

# Calculando os quartis
quartiles = np.percentile(alturas, [25, 50, 75])
quartiles
```

```python
"""
5) Em uma classe de estatística com 50 alunos, as notas obtidas formaram a
seguinte distribuição:
Notas        2 3 4 5 6 7 8 9 10
Nº de alunos 1 3 6 10 13 8 5 3 1
Calcule a nota média, a nota mediana e a nota modal.
"""

import numpy as np
from scipy import stats

# Dados das notas e número de alunos
notas = [2, 3, 4, 5, 6, 7, 8, 9, 10]
alunos_por_nota = [1, 3, 6, 10, 13, 8, 5, 3, 1]

# Expansão dos dados para cada nota
dados_expandidos = []
for nota, alunos in zip(notas, alunos_por_nota):
    dados_expandidos.extend([nota] * alunos)

print(dados_expandidos)

# Cálculo da média, mediana e moda
media = np.mean(dados_expandidos)
mediana = np.median(dados_expandidos)
moda = stats.mode(dados_expandidos)[0]

print(f'Média: {media}')
print(f'Mediana: {mediana}')
print(f'Moda: {moda}')
```
